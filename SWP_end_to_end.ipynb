```python
import os
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Function to download webpage content and extract links
def download_content_and_extract_links(url, base_url, visited_urls=set()):
    if url in visited_urls:
        return None, None
    visited_urls.add(url)
    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            # Extract text content
            text_content = soup.get_text()
            # Save text content to file
            filename = os.path.join("downloads", url.replace("/", "_").replace(":", "_") + ".txt")
            with open(filename, "w", encoding="utf-8") as file:
                file.write(text_content)
            # Extract links
            links = []
            for link in soup.find_all('a', href=True):
                href = link['href']
                if href.startswith("http"):
                    links.append(urljoin(base_url, href))
            return text_content, links
        else:
            return None, None
    except Exception as e:
        print(f"Error downloading content from {url}: {e}")
        return None, None

# Function to preprocess text
def preprocess_text(text):
    # Perform preprocessing steps such as tokenization, removing stopwords, etc.
    # Placeholder implementation
    return text.lower()

# Function to compute cosine similarity
def compute_similarity(content, documents):
    if not documents:
        return None
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform([content] + documents)
    similarity_scores = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:]).flatten()
    return similarity_scores

# Function to recursively download content from all sub webpages
def download_subpages(base_url):
    queue = [base_url]
    visited_urls = set()
    while queue:
        current_url = queue.pop(0)
        content, links = download_content_and_extract_links(current_url, base_url, visited_urls)
        if content:
            visited_urls.add(current_url)
            queue.extend(links)

# Example usage
if __name__ == "__main__":
    # URL of the base webpage
    base_url = "https://am.jpmorgan.com/us/en/asset-management/adv/"
    # Create the downloads directory if it doesn't exist
    if not os.path.exists("downloads"):
        os.makedirs("downloads")
    # Download content from all sub webpages
    download_subpages(base_url)
```

```python
import os
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Function to download webpage content
def download_content(url, visited_urls=set()):
    if url in visited_urls:
        return None
    visited_urls.add(url)
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        # Extract text content
        text_content = soup.get_text()
        return text_content
    else:
        return None

# Function to preprocess text
def preprocess_text(text):
    # Perform preprocessing steps such as tokenization, removing stopwords, etc.
    # Placeholder implementation
    return text.lower()

# Function to compute cosine similarity
def compute_similarity(content, documents):
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform([content] + documents)
    similarity_scores = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:]).flatten()
    return similarity_scores

# Example usage
if __name__ == "__main__":
    # URL to download content from
    url = "https://am.jpmorgan.com/us/en/asset-management/liq/products/jpmorgan-us-treasury-plus-money-market-fund-capital-4812c2239"
    # Download content from the URL
    content = download_content(url)
    if content:
        print("Downloaded content successfully:")
        print(content[:500])  # Print the first 500 characters of the downloaded content
        # Preprocess text
        preprocessed_content = preprocess_text(content)
        # Loop through the "downloads" directory
        downloaded_files = os.listdir("downloads")
        similar_documents = []
        for filename in downloaded_files:
            with open(os.path.join("downloads", filename), "r", encoding="utf-8") as file:
                text = file.read()
                preprocessed_text = preprocess_text(text)
                similar_documents.append(preprocessed_text)
        # Compute similarity
        similarity_scores = compute_similarity(preprocessed_content, similar_documents)
        # Recommend similar links
        recommended_links = [downloaded_files[i] for i in similarity_scores.argsort()[::-1][:3]]
        print("Recommended similar links:")
        for link in recommended_links:
            print(os.path.splitext(link)[0])
    else:
        print("Failed to download content.")
```
